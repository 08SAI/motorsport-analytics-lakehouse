```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

laps = spark.table("silver_lap_times")

w = Window.partitionBy("raceId").orderBy("milliseconds")

ranked_laps = laps.withColumn("rank", row_number().over(w))

ranked_laps.write.format("delta").mode("overwrite").option("overwriteSchema", True).saveAsTable("gold_fastest_lap")

ranked_laps.display()
```

```python
from pyspark.sql.functions import avg

consistency = laps.groupBy("driverId") \
                  .agg(avg("milliseconds").alias("avg_lap")) \
                  .orderBy("avg_lap")

consistency.write.format("delta").mode("overwrite").option("overwriteSchema", True).saveAsTable("gold_driver_consistency")
consistency.display()
```

```python
from pyspark.sql.functions import avg

w2 = Window.partitionBy("driverId").orderBy("lap")

laps = laps.withColumn("running_avg",
                       avg("milliseconds").over(w2))
```

```python
from pyspark.sql.functions import percent_rank

w3 = Window.partitionBy("raceId").orderBy("milliseconds")

laps = laps.withColumn("percentile",
                       percent_rank().over(w3))
```
