```sql
SELECT driverId, MIN(milliseconds) fastest_lap
FROM gold_fastest_lap
GROUP BY driverId
ORDER BY fastest_lap;
```

```sql
SELECT constructorId,
COUNT(*) * 100.0 / SUM(COUNT(*)) OVER() AS win_percent
FROM fact_race_results
GROUP BY constructorId;
```

```sql
SELECT r.circuitId,
AVG(g.milliseconds) avg_lap,
STDDEV(g.milliseconds) variability
FROM gold_fastest_lap g
JOIN bronze_races r ON g.raceId = r.raceId
GROUP BY r.circuitId;
```
