```python
%sql
-- Data quality checks for silver tables

-- Check 1: NULL milliseconds in lap times
SELECT 
  COUNT(*) as total_lap_times,
  SUM(CASE WHEN milliseconds IS NULL THEN 1 ELSE 0 END) as null_milliseconds_count,
  ROUND(SUM(CASE WHEN milliseconds IS NULL THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) as null_percentage
FROM silver_lap_times;
```

```python
%sql
-- Check 2: Duplicate drivers
SELECT 
  COUNT(DISTINCT driverId) as unique_drivers,
  COUNT(*) as total_driver_records,
  COUNT(*) - COUNT(DISTINCT driverId) as duplicate_records
FROM silver_drivers;
```
